```python
import xgboost
```

```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent, test_size=0.30,random_state=0)
```

```python
from xgboost import XGBRegressor
regressor=XGBRegressor(n_estimators=100, max_depth=3, alpha=1, learning_rate=0.05)
regressor.fit(X_train,Y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
r_score
```

```python
import pickle
filename="finalized_model_Xg boost.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_Xg boost.sav",'rb'))
result=loaded_model.predict([[35,28.88,0,1,1]])
result
```
